```python
arr = np.random.normal(0, 1, 1000000)
buckets = [arr]
num_buckets = 10
max_bucket_size = 20

iter = 0
while iter < 5:
    new_buckets = []
    flag = False
    
    for x in buckets:
        if len(x) > max_bucket_size:
            flag = True
            
            a = np.min(x)
            b = np.max(x)
            interval = (b-a)/num_buckets
            
            h = [[] for _ in range(num_buckets)]
            for z in x:
                m = int((z-a)/interval)
                if m == num_buckets:
                    h[num_buckets-1] += [z]
                else:
                    h[m] += [z]
            
            new_buckets += h
        else:
            new_buckets += [x]
            
    buckets = new_buckets[:]
    if flag is False:
        break
    
    iter += 1

curr_len = 0
merged_buckets = []
temp = []
for i in range(len(buckets)):
    if curr_len + len(buckets[i]) > max_bucket_size and len(temp) > 0:
        merged_buckets += [temp]
        temp = []
        curr_len = 0
        
    temp += buckets[i]
    curr_len += len(buckets[i])

if len(temp) > 0:
    merged_buckets += [temp]
```

```python
min_num_buckets = [float("Inf")]*len(buckets)
for i in range(len(buckets)):
    j = i
    s = 0
    while j >= 0:
        s += len(buckets[j])
        if s <= max_bucket_size:
            if j == 0:
                min_num_buckets[i] = 1
            else:
                min_num_buckets[i] = min(min_num_buckets[i], min_num_buckets[j-1]+1)
        else:
            break
        
        j -= 1
```
